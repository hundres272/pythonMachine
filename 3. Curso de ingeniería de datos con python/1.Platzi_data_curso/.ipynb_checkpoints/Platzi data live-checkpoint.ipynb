Solicitudes a la web **bold**

```python
import requests

response = requests.get('https://platzi.com/cursos/scikit/')

print(dir(response))
```

```python
print(response.text)
```

**Extracción de información del HTML**

```python
import bs4

soup = bs4.BeautifulSoup(response.text, 'html.parser')

courses_links = soup.select('.Material-name')
 
courses = [course.text for course in courses_links]
cont = 1

for course in courses:
    print(cont, course)
    cont += 1
```

```python
import requests

response = requests.get('https://diariodelsur.com.co/')

print(dir(response))
```

```python
import bs4

soup = bs4.BeautifulSoup(response.text, 'html.parser')

courses_links = soup.select('.field-item')
 
courses = [course.text for course in courses_links]

for course in courses:
    print(course)
```
